```python
import pandas as pd
```

```python
import numpy as np
```

```python
import os; import re
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
from sklearn.cross_validation import KFold
```

```python
from sklearn.grid_search import GridSearchCV
```

```python
import time
import datetime
```

```python
features = pd.read_csv('./_features/features.csv', index_col='match_id')
```

```python
dict_names = os.listdir("./_data/data/dictionaries")
```

###### create dict of dicts with object description

```python
dicts = dict()
for i,d in enumerate(dict_names):
    dicts[re.sub(r'(.*)\.csv',r'\1',d)] = pd.read_csv('./_data/data/dictionaries/%s' % d)
#print dicts
```

###### drop final attributes

```python
X = features.drop(features.columns[-6:],1)
```

###### found skips

```python
counts=X.count()
```

```python
skips=counts[counts.values < 97230].keys()
```

###### не в каждом матче FB происходит в первые 5 минут, не в каждом матче летучий курьер приобретается одной из команд в первые 5 минут

```python
skips[[0,-2]]
```

###### replace na items to ZERO

```python
X_na = X.fillna(0)
```

```python
X_ma=X.fillna(1000000000)
```

###### target var - radiant_win

```python
y=features.radiant_win
```

###### create cross validator

```python
cv = KFold(y.size,shuffle=True, n_folds=5)
```

###### сreate and fit GradientBoostingClassifier

```python
ranges=[10,20,30,60] #np.arange(10,20,10)
```

```python
times = pd.DataFrame({'n_estimators':ranges},columns=['n_estimators','time','score']).set_index('n_estimators')
```

```python
for n in [10,20,30,60]:
    params_grid={"n_estimators":[n]}
    gs = GridSearchCV(GradientBoostingClassifier(),param_grid=params_grid,cv=cv,scoring="roc_auc")
    start_time = datetime.datetime.now()
    #times['n_estimators:' + gs.param_grid.values()[0][0]] = 
    gs.fit(X_na,y)
    n_est=gs.param_grid.values()[0][0]
    et=(datetime.datetime.now() - start_time).seconds
    sc=gs.best_score_
    times.loc[n_est] = [et,sc]


#print 'Time elapsed:', datetime.datetime.now() - start_time
times
```

###### set subsample to 0.2

```python
ranges=np.arange(10,20,10)
```

```python
op_times = pd.DataFrame({'n_estimators':ranges},columns=['n_estimators','time','score']).set_index('n_estimators')
```

```python
for n in [10,20,30,60]:
    params_grid={"n_estimators":[n],'loss':['deviance'],'subsample':[0.1]}
    gs = GridSearchCV(GradientBoostingClassifier(),param_grid=params_grid,cv=cv,scoring="roc_auc")
    start_time = datetime.datetime.now()
    #times['n_estimators:' + gs.param_grid.values()[0][0]] = 
    gs.fit(X_na,y)
    n_est=gs.param_grid.values()[0][0]
    et=(datetime.datetime.now() - start_time).seconds
    sc=gs.best_score_
    
    op_times.loc[n_est] = [et,sc]

#print 'Time elapsed:', datetime.datetime.now() - start_time
op_times
```

###### with mean in skips

```python
ranges=np.arange(10,20,10)
```

```python
m_times = pd.DataFrame({'n_estimators':ranges},columns=['n_estimators','time','score']).set_index('n_estimators')
```

```python
for n in ranges:
    params_grid={"n_estimators":[n],'loss':['exponential'],'subsample':[0.1],'learning_rate':[0.2]}
    gs = GridSearchCV(GradientBoostingClassifier(),param_grid=params_grid,cv=cv,scoring="roc_auc")
    start_time = datetime.datetime.now()
    #times['n_estimators:' + gs.param_grid.values()[0][0]] = 
    gs.fit(X_ma,y)
    n_est=gs.param_grid.values()[0][0]
    et=(datetime.datetime.now() - start_time).seconds
    sc=gs.best_score_
    
    m_times.loc[n_est] = [et,sc]

#print 'Time elapsed:', datetime.datetime.now() - start_time
m_times
```
